```python
import csv

with open("occupations-truncated.csv",'r') as csv_data:
    csv_reader = csv.reader(csv_data, delimiter=',')
    data = [file for file in csv_reader]
data
# Here we can see that the the data is organized in the following way. 
#Occupation code index 0
#Occupation title index 1
#Level index 2
#Employment index 3
#Annual mean wage index 4
```

```python
# I created a dictionary in which the occupation is the key and the level and wage the values stored in a list
occupations = {line[1]:[line[2],line[4]]for line in data if 'computer'in line[1] or 'Computer'in line[1] or 'Data'in line[1] or'data'in line[1]}
occupations
```

```python
occupation_detailed = {k:v[1].strip('$').replace(',','')for k,v in occupations.items() if v[0] =='detail'}
occupation_detailed
```

```python
list_of_salaries = [int(v) for k,v in occupation_detailed.items()]

min_value = min(list_of_salaries)
max_value = max(list_of_salaries)
final_occupations = [(k,v) for k,v in occupation_detailed.items() if int(v) == min_value or int(v) ==max_value]
```
