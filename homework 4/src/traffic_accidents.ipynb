### 8 - what are the distributions of accidents based on the geo location (latitude & longitude)?



```python
# this one is hard, but very cool
# first group locations by borough. 
# then plot each one on top of the other with differen colors
# R = 6371, radius of earth


geo_location_borough = pd.DataFrame(data, columns = ['BOROUGH','LATITUDE','LONGITUDE'])
bronx = geo_location_borough[geo_location_borough['BOROUGH']=='BRONX']
brooklyn = geo_location_borough[geo_location_borough['BOROUGH']=='BROOKLYN']
staten_island = geo_location_borough[geo_location_borough['BOROUGH']=='STATEN ISLAND']
queens = geo_location_borough[geo_location_borough['BOROUGH']=='QUEENS']
manhattan = geo_location_borough[geo_location_borough['BOROUGH']=='MANHATTAN']


plt.scatter(x='LATITUDE' ,y='LONGITUDE',data=bronx,alpha = 0.1,color='r')
plt.scatter(x='LATITUDE' ,y='LONGITUDE',data=brooklyn,alpha = 0.1,color='b')
plt.scatter(x='LATITUDE' ,y='LONGITUDE',data=staten_island,alpha = 0.1,color='y')
plt.scatter(x='LATITUDE' ,y='LONGITUDE',data=queens, alpha = 0.1,color='g')
plt.scatter(x='LATITUDE' ,y='LONGITUDE',data=manhattan, alpha = 0.1,color='c')
plt.xlabel("LATITUDE")
plt.ylabel("LONGITUDE")

# x = R * cos (latitude) * cos (longitude)
# y = R * cos (latitude) * sin (longitude)
plt.ylim(-74.4,-73.4,)
plt.xlim(40.4,41)
plt.show()
```

Firs I isolated the columns of borough, latitude and longitude into a new dataframe. Then I separated the data from that data from on five different dataframes, each one in respective to a different borough.
Then I plotted each one separately, which turned out to be the NYC map but in a different orientation.


### 9 - shows the covariance between each pair of the columns

I chose the Zip code, casualties(injured people+ people killed on accidents) and latidude, because they were the easiest numerical data to work with.
I make sense to get the covarience in between the zip and casualties, because there most be some region where acciendents occur more often and others that barely occur.

```python
data['casualties'] = data['NUMBER OF PERSONS INJURED']+data['NUMBER OF PERSONS KILLED']
date_and_zip = data[['ZIP CODE','casualties']]
latitude_casualties = data[['LATITUDE','casualties']]
print(date_and_zip.cov())
print(latitude_casualties.cov())
```

First I created a new column called casualties, from the columns NUMBER OF PERSONS INJURED and NUMBER OF PERSONS KILLED. 
Then I isolated this column with the Zip code column and then with the Latitude column on separate data frames.
